```python
def find_outliers(dataset, column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dataset[(dataset[column] < lower_bound) | (dataset[column] > upper_bound)]
    return outliers

numerical_columns = ['Salary','10percentage', 'ComputerScience', 
                     'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 
                     'conscientiousness', 'agreeableness', 'extraversion', 
                     'nueroticism', 'openess_to_experience']

outliers_dict = {}
for col in numerical_columns:
    outliers_dict[col] = find_outliers(data, col)

# Display the number of outliers for each numerical column
for col, outliers in outliers_dict.items():
    print(f"Number of outliers in {col}: {outliers.shape[0]}")


num_plots = len(numerical_columns)
num_cols = 4  # Number of columns in the subplot grid
num_rows = -(-num_plots // num_cols)  # Calculate the number of rows needed

plt.figure(figsize=(25, num_rows * 5))

for idx, col in enumerate(numerical_columns):
    plt.subplot(num_rows, num_cols, idx + 1)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    outliers = outliers_dict[col]
   

plt.tight_layout()
plt.show()
```

```python
# Convert date columns to datetime format
dataset['DOJ'] = pd.to_datetime(dataset['DOJ'], errors='coerce')
dataset['DOL'] = pd.to_datetime(dataset['DOL'], errors='coerce')

# Convert date columns to Unix timestamps
dataset['DOJ'] = dataset['DOJ'].apply(lambda x: x.timestamp() if not pd.isnull(x) else x)
dataset['DOL'] = dataset['DOL'].apply(lambda x: x.timestamp() if not pd.isnull(x) else x)


numerical_columns = ['Salary', '10percentage', 'ComputerScience', 
                     'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 
                     'conscientiousness', 'agreeableness', 'extraversion', 
                     'nueroticism', 'openess_to_experience']

num_plots = len(numerical_columns)
num_cols = 4  # Number of columns in the subplot grid
num_rows = -(-num_plots // num_cols)  # Calculate the number of rows needed

plt.figure(figsize=(20, num_rows * 5))

for idx, col in enumerate(numerical_columns):
    plt.subplot(num_rows, num_cols, idx + 1)
    sns.histplot(dataset[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Observations
    print(f"Observations for {col}:")
    print(f"- Mean: {dataset[col].mean()}")
    print(f"- Median: {dataset[col].median()}")
    print(f"- Standard Deviation: {dataset[col].std()}")
    print(f"- Skewness: {dataset[col].skew()}")
    print(f"- Kurtosis: {dataset[col].kurtosis()}")
    print("\n")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to plot and print observations
def plot_and_observe(column):
    plt.figure(figsize=(10, 5))
    sns.countplot(x=dataset[column], order=dataset[column].value_counts().index[:10])  # Limit to top 10 categories for readability
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

    # Observations
    value_counts = dataset[column].value_counts()
    print(f"Observations for {column}:")
    print(value_counts)
    print("\n")

# Plot and observe each categorical column
for i in dataset.columns:
    if dataset[i].dtype == 'object':
        plot_and_observe(i)
```

### Stacked Bar Plot: Gender vs. Specialization

```python
specialization_columns = ['ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']
dataset['Specialization'] = dataset[specialization_columns].idxmax(axis=1)

ct = pd.crosstab(dataset['Gender'], dataset['Specialization'])

ct.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Stacked Bar Plot: Gender vs. Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Insight:**
'''The stacked bar plot illustrates the distribution of specializations among different genders within the dataset. 
It reveals which specializations are more popular among male and female candidates. 
By observing the colors and hights of the bars, one can infer trends in gender preferences for various engineering fields,
highlighting potential gender disparities in specialization choices.'''

#### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you

```python
from scipy import stats
designation = ['programming Analyst', 'software engineer', 'hardware engineer', 'associate engineer']
filter = dataset[dataset['Designation'].isin(designation)]
salary_data=filter['Salary']
mean_salary=2.75 * 100000
t_stat,p_value = stats.ttest_1samp(salary_data,mean_salary)
print(f'Mean salary of selected role :{salary_data.mean(): .2f}')
print(f'Claimed Mean Salary :{mean_salary: .2f}')
print(f'T-statistic :{t_stat: .2f}')
print(f'P-value :{p_value: .4f}')

avg=0.05
if p_value < avg:
    print('There is a significant relationship between gender and specialization preference.')
else:
    print('There is no significant relationship between gender and specialization preference.')
```

#### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Specify the specialization column for Computer Science
specialization_columns = ['ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']
dataset['Specialization'] = dataset[specialization_columns].idxmax(axis=1)

# Create a contingency table
ct = pd.crosstab(dataset['Gender'], dataset['Specialization'])

# Display the contingency table
print(ct)

# Plotting the relationship
ct.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Stacked Bar Plot: Gender vs. Specialization')
plt.xlabel('Gender')
plt.ylabel('Count of Specialization')
plt.legend(title='Specialization')
plt.show()
```
